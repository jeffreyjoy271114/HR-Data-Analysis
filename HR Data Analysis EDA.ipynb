```python
hrdata.groupby('Gender').count()['Attrition'].plot(kind = 'bar')
plt.title('Gender with Attrition')
plt.ylabel('Attrition Rate')
```

```python
hrdata.groupby(['Marital Status']).count()['Attrition'].plot(kind = 'bar')
plt.title('Marital Status with Attririon')
plt.ylabel('Attrition Rate')
```

```python
hrdata.groupby(['Department']).count()['Attrition'].plot(kind = 'bar')
plt.title('Attrition with Department')
plt.ylabel('Attrition Rate')
```

```python
hrdata.groupby(['Job Role']).count()['Attrition'].sort_values(ascending = False).plot(kind = 'bar')
plt.title('Attrition with Job Role')
plt.ylabel('Attrition Rate')
```

```python
hrdata.groupby(['Education', 'Attrition']).count()['Job Role'].plot(kind = 'bar', stacked = True)
plt.title('Eduaction with Job Role')
plt.ylabel('Job roles According to education')
plt.legend()
```

```python
# Create the age bins
age_bins = range(18, 61, 5)
labels = ['{}-{}'.format(start, start+4) for start in age_bins[:-1]]

# Assign the age bins to a new column in the DataFrame
hrdata['Age Group'] = pd.cut(hrdata['Age'], bins=age_bins, labels=labels, right=False)

# Group by 'Gender' and 'Age Group', and count the number of 'Attrition' occurrences
grouped_data = hrdata.groupby(['Gender', 'Age Group']).count()['Attrition'].sort_values(ascending = False)

# Plot the grouped data as a bar plot
grouped_data.plot(kind='bar')
plt.xlabel('Gender and Age Group')
plt.ylabel('Count')
plt.title('Count of Attrition by Gender and Age Group')
plt.show()
```
